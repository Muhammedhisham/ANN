```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('weatherAUS.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
# data['Date']=label_encoder.fit_transform(data['Date'])
data['Location']=label_encoder.fit_transform(data['Location'])
data['WindGustDir']=label_encoder.fit_transform(data['WindGustDir'])
data['WindDir9am']=label_encoder.fit_transform(data['WindDir9am'])
data['RainTomorrow']=label_encoder.fit_transform(data['RainTomorrow'])
data['RainToday']=label_encoder.fit_transform(data['RainToday'])
data['WindDir3pm']=label_encoder.fit_transform(data['WindDir3pm'])
```

```python
data.head()
year=[]
month=[]
day=[]

data['Date']
for i in data['Date']:
  i=i.split('-')
  year.append(int(i[0]))
  month.append(int(i[1]))
  day.append(int(i[2]))

data['Year']=year
data['Month']=month
data['Day']=day
```

```python
data.drop(['Date'],axis=1,inplace=True)
```

```python
data
```

```python
y=data['RainTomorrow']
x=data.drop('RainTomorrow',axis=1)
```

```python
y
```

```python
x
```

```python
data['RainToday'].unique()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import binary_crossentropy
```

```python
model=Sequential()
model.add(Flatten(input_shape=(x_train.shape[1],)))
model.add(Dense(100,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
x_train
y_train.dtype
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
print(x_train.dtypes)
print(y_train.dtypes)
print(x_test.dtypes)
print(y_test.dtypes)
```

```python
data.info()
```

```python
data['WindDir3pm']
pd.set_option('display.max_columns',None)
```

```python
data.head()
```

```python
h=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=20)
```

```python
h.history
```

```python
t_a=h.history['accuracy']
v_a=h.history['val_accuracy']
t_l=h.history['loss']
v_l=h.history['val_loss']
epoch=h.epoch
plt.plot(epoch,t_a,label=['accouracy'])
plt.plot(epoch,v_a,label=['validation accouracy'])
plt.title('accouracy LC')
plt.xlabel('Training accouracy')
plt.ylabel('validation accouracy')
plt.legend()

```

```python
plt.plot(epoch,t_l,label=['loss'])
plt.plot(epoch,v_l,label=['validation loss'])
plt.title('loss LC')
plt.xlabel('Training loss')
plt.ylabel('validation loss')
plt.legend()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
# Convert probabilities to binary labels using a threshold
prediction_binary = (prediction > 0.5).astype(int)

# Now calculate metrics with binary labels
accuracy = accuracy_score(y_test, prediction_binary)
precision = precision_score(y_test, prediction_binary)
recall = recall_score(y_test, prediction_binary)
f1 = f1_score(y_test, prediction_binary)
conf_matrix = confusion_matrix(y_test, prediction_binary)

# Print the metrics
print("Confusion Matrix:")
print(conf_matrix)
print("Accuracy:", accuracy)
print("Precision", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
